```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
raw_data=pd.read_csv("order_items.csv")
raw_data
```

```python
data=raw_data.groupby(['seller_id'],as_index=False).count().sort_values(by=['order_id'])
data
```

```python
##df.loc[(df['col1'] == value) & (df['col2'] < value)]
outsiders_sellers=data.loc[data['order_id']==1]
outside_sellers_id=outsiders_sellers['seller_id']
##outside_sellers_id

leaders_sellers=data.loc[data['order_id']>=300]
leaders_sellers_id=leaders_sellers['seller_id']
leaders_sellers_id
```

```python
outside_orders=[]
for ind in raw_data.index:
    if(ind in outside_sellers_id.index):
        outside_orders.append(raw_data['order_id'][ind])

##outside_orders

leader_orders=[]
for ind in raw_data.index:
    if(ind in leaders_sellers_id.index):
        leader_orders.append(raw_data['order_id'][ind])
leader_orders
```

```python
order_data=pd.read_csv('orders.csv')
order_data
```

```python
outside_customer_id=[]

for ind in order_data.index:
    if(order_data['order_id'][ind] in outside_orders):
        outside_customer_id.append(order_data['customer_id'][ind])
##outside_customer_id
    
leader_customer_id=[]
for ind in order_data.index:
    if(order_data['order_id'][ind] in leader_orders):
        leader_customer_id.append(order_data['customer_id'][ind])
leader_customer_id
```

```python
customer_data=pd.read_csv('customers.csv')
customer_data
```

```python
outside_cities=[]
for ind in customer_data.index:
    if(customer_data['customer_id'][ind] in outside_customer_id):
        outside_cities.append(customer_data['customer_city'][ind])
##outside_cities

leader_cities=[]
for ind in customer_data.index:
    if(customer_data['customer_id'][ind] in leader_customer_id):
        leader_cities.append(customer_data['customer_city'][ind])
leader_cities
```

```python
unique_cities=set(outside_cities)
results=pd.DataFrame(unique_cities)
unique_amounts=[]
for a in unique_cities:
    unique_amounts.append(outside_cities.count(a))
results['Amounts']=unique_amounts
##results
leader_unique_cities=set(leader_cities)
leader_results=pd.DataFrame(leader_unique_cities)
leader_unique_amounts=[]
for a in leader_unique_cities:
    leader_unique_amounts.append(leader_cities.count(a))

leader_results['Amounts']=leader_unique_amounts
leader_results
```
